#### Explore Values of HS2 Backscatter Data Over Time

This Notebook is part of the auv-python project (https://github.com/mbari-org/auv-python).

```python
# Define function for plotting percentiles given a dataframe and variable name
import matplotlib.pyplot as plt

def plot_percentiles(df, variable, ylim=(0, 0.03)):
    plt.figure(figsize=(12, 6))
    plt.plot(df['activity__startdate'], df['p010'], label='1.0th Percentile', color='blue')
    plt.plot(df['activity__startdate'], df['p025'], label='2.5th Percentile', color='orange')
    plt.plot(df['activity__startdate'], df['p975'], label='97.5th Percentile', color='green')
    plt.plot(df['activity__startdate'], df['p990'], label='99.0th Percentile', color='red')
    plt.title(f'Percentiles of {variable} Over Time')
    plt.xlabel('Time (activity__startdate)')
    plt.ylabel(f'{variable}')
    plt.ylim(ylim)  # Adjust these limits based on your data
    # Spread out the x-axis labels, say every 10th label
    plt.xticks(df['activity__startdate'][::10], rotation=45, ha='right')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
    plt.show()
```

```python
import pandas as pd

url = "https://stoqs.shore.mbari.org/stoqs_all_dorado/api/activityparameter.csv?"
url += "activity__platform__name=dorado&parameter__name=hs2_bb420%20(m-1)"
df = pd.read_csv(url)
plot_percentiles(df, 'hs2_bb420 (m-1)')
```
