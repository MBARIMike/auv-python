The urls in the pooch.retrieve() calls  below were generated by going to https://stoqs.shore.mbari.org/stoqs_auv_compare/ and clicking the buttons of the Measured Parameters to be included in the Parquet file. Then clicking the "Measured Parameter Data Access" section and clicking the "Estimate requirements" button to verify that the estimated values are within the available values of the server. 

````python
import pandas as pd
import pooch
````

````python
# All of the Gulper measurements from a Campaign
gulper_file = pooch.retrieve(
    url="https://stoqs.shore.mbari.org/stoqs_auv_compare/api/measuredparameter.parquet?measurement__instantpoint__activity__platform__name=dorado_Gulper&collect=name&include=activity__name",
    known_hash="7e925a33e44e7047401180be0816fb8611a06171a723c67d8d79d504f5928edf",
)
df = pd.read_parquet(gulper_file)
df.describe()
````

````python
# Measurements from a single Gulper
gulp6_file = pooch.retrieve(
    url="https://stoqs.shore.mbari.org/stoqs_auv_compare/api/measuredparameter.parquet?measurement__instantpoint__activity__name__contains=dorado_2016.270.00_1S.nc_Gulper_6&collect=name&include=activity__name",
    known_hash="f0da1be5a70d0fb502e03daa0e72c66e2be623a4e8a0d3ad99bd4758b35d5f0b",
)
df_gulp6 = pd.read_parquet(gulp6_file)
df_gulp6.describe()
````
